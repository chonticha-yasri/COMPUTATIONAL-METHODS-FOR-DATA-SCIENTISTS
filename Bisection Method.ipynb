Bisection Method

```python
import math
```

```python
x = 2
print(((math.e)**x)-(3*x))
```

```python
x = []
fc1 = []
max_iterations = 100
A = 1
B = 0
for i in range(max_iterations):
        C = (A+B)/2
        x = x + [C]
        
        fc = ((math.e)**x[i])-(3*x[i])
        fc1.append(fc)
        
        print('A%d= %.8f'%(i+1,A),'   B%d= %.8f'%(i+1,B),'    C%d =%.8f'%(i+1,C),'      fc%d=  %.8f'%(i+1,fc))
        if fc1[i]>0:
            B=C
        elif fc1[i]<0:
            A=C
        if abs(fc1[i]-0) < 10**-8:
            print()
            print('exact = ',C)
            print('fc%d = %.10f'%(i+1,fc))
            break
        
```

หาผลเฉลยของโดยวิธีวางตัวผิดที่หรือวิธีแก้ตำแหน่งผิด

```python
x = []
fc1 = []
max_iterations = 100
A = 1
B = 0
for i in range(max_iterations):
        fb = ((math.e)**B)-(3*B)
        fa = ((math.e)**A)-(3*A)
        C = A-(((B-A)/(fb-fa))*fa)
        x = x + [C]
        
        fc = ((math.e)**x[i])-(3*x[i])
        fc1.append(fc)
        
        print('A%d= %.8f'%(i+1,A),'   B%d= %.8f'%(i+1,B),'    C%d =%.8f'%(i+1,C),'      fc%d=  %.8f'%(i+1,fc))
        if fc1[i]>0:
            B=C
        elif fc1[i]<0:
            A=C
        if abs(fc1[i]-0) < 10**-8:
            print()
            print('exact = ',C)
            print('fc%d = %.10f'%(i+1,fc))
            break
```
