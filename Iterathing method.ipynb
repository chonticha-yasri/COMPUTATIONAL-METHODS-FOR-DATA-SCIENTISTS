```python
x=[5]
max_iterations = 100
print('x0 = ',x[0])
for i in range(max_iterations):
        c = (3*x[i]+4)**(1/2)
        x = x + [c]
        print('x%d ='%(i+1),c)
        if x[i]-x[i+1] < 10**-4:
            break
```

```python
x=[30]
max_iterations = 100
print('x0 = ',x[0])
for i in range(max_iterations):
        c = ((485/(x[i]**2)-(5*(x[i]**2)))-3)
        x = x + [c]
        print('x%d ='%(i+1),c)
        if abs(x[i]-x[i+1]) < 10**-4:
            break
```

```python
x=[23]
max_iterations = 100
print('x0 =',x[0])
for i in range(max_iterations):
        c = (((485*x[i])-5)-(3*(x[i]**2)))**(1/3)
        x = x + [c]
        print('x%d ='%(i+1),c)
        if abs(x[i]-x[i+1]) < 10**-10:
            break
```

```python
import matplotlib.pyplot as plt 
import numpy as np 
  
x = np.linspace(-50, 50, 200) 
  
fig = plt.figure(figsize = (14, 8)) 
  
# Plot y = cos(x) 
y1 = ((x**3)+(3*(x**2)))/485
plt.plot(x, y1, 'b', label ='y1 = ((x**3)+(3*(x**2)))/485') 
  
# Plot degree 2 Taylor polynomial 
y2 = x + (1/97)
plt.plot(x, y2, 'r-.', label ='y2 = x + (1/97)') 
  
  
plt.legend() 
plt.grid(True, linestyle ='-') 
plt.xlim([-50, 50]) 
plt.ylim([-50, 50]) 
  
plt.title('x^3+3x^3-485x+5 = 0') 
plt.xlabel('x-axis') 
plt.ylabel('y-axis') 
  
# Show plot 
plt.show() 
# plt.savefig('fname.pdf')
```
